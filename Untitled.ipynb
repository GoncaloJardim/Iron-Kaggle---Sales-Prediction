```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import pickle

gb_reg = pickle.load(open("pickle_ironkaggle.p","rb"))
```

```python
testing = pd.read_csv("validation_for_students.csv")
testing.head(5)
```

```python
testing.shape
```

```python
testing_open = testing[testing["Open"]==1]
```

```python
testing_open
```

```python
testing_open.shape
```

```python
testing_open = pd.concat([testing_open, pd.get_dummies(testing_open['Day_of_week'])], axis=1)

testing_open["Date"] = pd.to_datetime(testing_open["Date"])

testing_open = pd.concat([testing_open, pd.get_dummies(testing_open["State_holiday"])], axis=1)


testing_open["date_year"] =testing_open["Date"].dt.year
testing_open["date_month"] = testing_open["Date"].dt.month
testing_open["date_week"] = testing_open["Date"].dt.weekofyear
testing_open["date_day"] = testing_open["Date"].dt.day
testing_open = pd.concat([testing_open, pd.get_dummies(testing_open["date_year"], prefix="Year")], axis=1)
testing_open= pd.concat([testing_open, pd.get_dummies(testing_open["date_month"], prefix="Month")], axis=1)
testing_open= pd.concat([testing_open, pd.get_dummies(testing_open["date_week"], prefix="Week")], axis=1)
testing_open= pd.concat([testing_open, pd.get_dummies(testing_open["date_day"], prefix="Day")], axis=1)
```

```python
xgb_reg = pickle.load(open("pickle_ironkaggle.p","rb"))
```

```python
testing_open
```

```python
#testing = testing.drop(columns=["Day_of_week", "Date", "State_holiday"], axis=1) ###
```

```python
normalizer = MinMaxScaler()

testing_normalized = pd.DataFrame(normalizer.fit_transform(testing_open.drop(columns=["Day_of_week", "Date", "State_holiday", 'True_index'], axis=1)), columns = testing_open.drop(columns=["Day_of_week", "Date", "State_holiday", 'True_index'], axis=1).columns)
```

```python
testing_normalized['True_index'] = testing_open['True_index']
```

```python
sales_pred = xgb_reg.predict(testing_normalized)
```

```python
sales_pred
```

```python
predictions = pd.Series(sales_pred, name = 'predictions')
predictions
```

```python
true_index = testing_open['True_index'].reset_index().drop(columns = ['index'])
true_index
```

```python
open_pred = pd.concat([true_index, predictions], axis = 1)
open_pred
```
